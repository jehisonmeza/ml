```python
import numpy as np
import matplotlib.pyplot as plt



import scipy
import pandas as pd

import h5py
from PIL import Image


from skimage.transform import rescale, resize, downscale_local_mean


def load_dataset():
     
    matriz=np.array([])
    
    for m in range(1,1639):
         
        imagen=Image.open("C:/imagenes/"+str(m)+".jpg")
         
        pixel_arr=np.array(imagen)
        
        print("--------------arreglo de la imagen "+str(m)+"----------------------------")     
               
        my_image = scipy.misc.imresize(pixel_arr, size=(100,100))

       
        
        print("---------------------------- con X-----------------------------")
  
        print("----------------------------agregar imagen a matriz sin organizada- "+str(m)+"-----------------------")          
        matriz=np.append(matriz,my_image)
        
    print("--------------------------------creando matriz final-------------------------------")        

    #segmentar todo arreglo (Xn) que entre en (X)
   
    im=matriz.reshape(m,100,100,3)
 
 
    
    print("-------------------------------------sacar imagen/arreglo del la percistencia ---------------------------")

    f=h5py.File("c:/data/train_data.hdf5", "w")
    # crear el dataset
    dset=f.create_dataset('train_data', data=im.astype(int))

    dataset1 = h5py.File('C:/data/train_data.hdf5', "r")
    #leer los elementos del dataset
    D = np.array(dataset1['train_data'][:])
    print(D.shape)
    plt.imshow(D[0])
    plt.show
    
    
  

 
     
                
        
        
       
      
```

```python
load_dataset() 
```
